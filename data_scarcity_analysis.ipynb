```python
import seaborn
import matplotlib.pyplot as plt
import os
import pickle
import pandas as pd
```

```python
def construct_graph_coarse(source, target):
    methods = ['TGONLY', 'SCTGJL', 'SCTGCL', 'FA', 'LA', 'IW']
    datasize = list(range(100, 1001, 100))
    folder = '../models/DS_{}_{}_{}_coarse/'
    f1_scores = {}
    for method in methods:
        results = pickle.load(open(os.path.join(folder.format(source, target, method), 'results.pkl'), 'rb'))
        f1_scores[method] = []
        for size in datasize:
            f1_scores[method].append(round((results[size]['target']['overall_f1']) * 100, 2))
    df = pd.DataFrame.from_dict(f1_scores)
    df.index = datasize
    plot = seaborn.lineplot(data=df)
    plot.set_ylim(50,98)
    plot.legend(bbox_to_anchor=(1.01, 1), borderaxespad=0)
    plot.set_xlabel("Number of labeled examples from target domain", fontsize = 16)
    plot.set_ylabel("F1 Score", fontsize = 16)
    plot.set_title("Transfer from {} to {} (Coarse NER)".format(source, target), fontsize=15.5)
    plot.set_xticklabels(plot.get_xticks(), fontsize = 14)
    plot.set_yticklabels(plot.get_yticks(), fontsize = 14)
    plt.setp(plot.get_legend().get_texts(), fontsize='14')
    plt.tight_layout()
    plt.savefig("{}_{}_coarse.jpg".format(source, target), format='jpg')
```

```python
construct_graph_coarse('conll2003', 'i2b22006')
```

```python
construct_graph_coarse('conll2003', 'i2b22010')
```

```python
construct_graph_coarse('conll2003', 'i2b22014')
```
